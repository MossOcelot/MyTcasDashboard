```python
import numpy as np
import pandas
import json
```

```python
with open('datasets/data4.json', encoding="utf8") as json_file:
    data = json.load(json_file)
```

```python
print(data['university'])
university = data["university"]
length = len(data['details'])
data_dict = {'university': [], 'major': [], 'course': [], 'ชื่อหลักสูตร': [], 'ชื่อหลักสูตรภาษาอังกฤษ': [], 'ประเภทหลักสูตร': [], 'วิทยาเขต': [], 'ค่าใช้จ่าย': [], 'อัตราการสำเร็จการศึกษา': [],
        'รอบ 1 Portfolio': [], 'รอบ 2 Quota': [], 'รอบ 3 Admission': [], 'รอบ 4 Direct Admission': []}
for i in range(length):
    detail = data['details'][i]

    major = detail['major']
    course_count = len(detail['course'])
    for j in range(course_count):
        detail_course = detail['course'][j]

        course_name = detail_course['course_name']
        status_rate = False
        status_price = False
        status_course_eng = False
        for de in detail_course['details_course'][0]:
            if de[0] != '':
                print(de[0])
                if de[0] != 'อัตราการได้งานทำ' and de[0] != "ค่ามัธยฐานเงินเดือน":
                    data_dict[de[0]].append(de[1])
            if de[0] == 'อัตราการสำเร็จการศึกษา':
                status_rate = True
            if de[0] == 'ค่าใช้จ่าย':
                status_price = True
            if de[0] == 'ชื่อหลักสูตรภาษาอังกฤษ':
                status_course_eng = True
        if not status_rate:
            data_dict['อัตราการสำเร็จการศึกษา'].append('')
        
        if not status_price:
            data_dict['ค่าใช้จ่าย'].append('')

        if not status_course_eng:
            data_dict['ชื่อหลักสูตรภาษาอังกฤษ'].append('')

        data_dict['university'].append(university)
        data_dict['major'].append(major)
        data_dict['course'].append(course_name)
        
        
        
```

```python
import os

folders = os.listdir('datasets')
sum_dataframe = pandas.DataFrame.from_dict({'university': [], 'major': [], 'course': [], 'ชื่อหลักสูตร': [], 'ชื่อหลักสูตรภาษาอังกฤษ': [], 'ประเภทหลักสูตร': [], 'วิทยาเขต': [], 'ค่าใช้จ่าย': [], 'อัตราการสำเร็จการศึกษา': [],
            'รอบ 1 Portfolio': [], 'รอบ 2 Quota': [], 'รอบ 3 Admission': [], 'รอบ 4 Direct Admission': []})

for folname in folders:
    with open(f'datasets/{folname}', encoding="utf8") as json_file:
        data = json.load(json_file)

    print(data['university'])
    university = data["university"]
    length = len(data['details'])
    data_dict = {'university': [], 'major': [], 'course': [], 'ชื่อหลักสูตร': [], 'ชื่อหลักสูตรภาษาอังกฤษ': [], 'ประเภทหลักสูตร': [], 'วิทยาเขต': [], 'ค่าใช้จ่าย': [], 'อัตราการสำเร็จการศึกษา': [],
            'รอบ 1 Portfolio': [], 'รอบ 2 Quota': [], 'รอบ 3 Admission': [], 'รอบ 4 Direct Admission': []}
    for i in range(length):
        detail = data['details'][i]

        major = detail['major']
        course_count = len(detail['course'])
        for j in range(course_count):
            detail_course = detail['course'][j]

            course_name = detail_course['course_name']
            status_rate = False
            status_price = False
            status_course_eng = False
            for de in detail_course['details_course'][0]:
                if de[0] != '':
                    print(de[0])
                    if de[0] != 'อัตราการได้งานทำ' and de[0] != "ค่ามัธยฐานเงินเดือน" and de[0] != 'สาขาวิชา':
                        data_dict[de[0]].append(de[1])
                if de[0] == 'อัตราการสำเร็จการศึกษา':
                    status_rate = True
                if de[0] == 'ค่าใช้จ่าย':
                    status_price = True
                if de[0] == 'ชื่อหลักสูตรภาษาอังกฤษ':
                    status_course_eng = True
            if not status_rate:
                data_dict['อัตราการสำเร็จการศึกษา'].append('')
            
            if not status_price:
                data_dict['ค่าใช้จ่าย'].append('')

            if not status_course_eng:
                data_dict['ชื่อหลักสูตรภาษาอังกฤษ'].append('')

            data_dict['university'].append(university)
            data_dict['major'].append(major)
            data_dict['course'].append(course_name)
        
    for k in data_dict.keys():
        print(k, len(data_dict[k]))

    df = pandas.DataFrame.from_dict(data_dict)
    sum_dataframe = pandas.concat([sum_dataframe, df], ignore_index = True)
```
