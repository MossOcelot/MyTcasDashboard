```python
import pandas
import numpy as np
```

```python
df = pandas.read_csv('tcas_dataset.csv')
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

```python
import re
def extract_numbers(text):
    if isinstance(text, str):
        numbers = re.findall(r'\d+', text)
        return int(numbers[0]) if numbers else 0
    return 0


df['รอบ 1 Portfolio'] = df['รอบ 1 Portfolio'].apply(extract_numbers)
df['รอบ 2 Quota'] = df['รอบ 2 Quota'].apply(extract_numbers)
df['รอบ 3 Admission'] = df['รอบ 3 Admission'].apply(extract_numbers)
df['รอบ 4 Direct Admission'] = df['รอบ 4 Direct Admission'].apply(extract_numbers)

df.to_csv('tcas_dataset.csv')
```
